```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

```python
plt.subplot(2, 1, 1)
plt.plot([2, 1, 3], label='Item 1')
plt.title('Subplot 1')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot([1, 3, 2], label='Item 2')
plt.title('Subplot 2')
plt.legend()

plt.tight_layout()
```
