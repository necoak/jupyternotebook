```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# R言語のggplotスタイル
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot([0, 1, 2, 10], [1, 2, 3, 11], label='data1')
ax.plot([0, 1, 2, 10], [0, 2, 4, 20], label='data2')
ax.legend() # 凡例
ax.set_xticks(range(0,11)) # X軸の軸項目
ax.set_yticks(range(0, 21, 5)) # Y軸の軸項目
ax.set_xlabel('X label', style='italic', size=16) # X軸ラベル
ax.set_ylabel('Y label', style='italic', size=20) # Y軸ラベル
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots()
ax.plot([0, 1, 2], [1, 2, 3])
ax.plot([0, 2, 4])
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Item1', 'Item2', 'Item3']) # 軸項目を数値から項目名にすることができる
```

```python
plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots()
ax.plot([1, 2, 3])
ax.plot([0, 2, 4]);
```
